```python
from simulation import *
from aesthetics import *
import matplotlib.pyplot as plt
from scipy.stats import norm
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
def gaussian(x, mu, sig):
    return np.exp(-np.power(x - mu, 2.) / (2 * np.power(sig, 2.)))    
```

```python
bins = range(60)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(bins, [gaussian(i, 30, 10) for i in bins], c='b', label='Landscape 1')
ax.plot(bins, [gaussian(i, 40, 2) for i in bins], c='r', label='Landscape 2')
ax.set_ylabel('Probability')
ax.set_xlabel('Coordinate')
ax.legend(loc=2)
fetching_plot(fig)
```

```python
this = simulation()
this.unbound_population = [gaussian(i, 30, 10) for i in bins]
this.bound_population = [gaussian(i, 40, 6) for i in bins]
this.simulate(pka_md_data = False)
```

```python
this.plot_energy()
this.plot_ss()
this.plot_flux()
```

```python
def population_2(offset):
    start = 29 + offset
    end = 29 - offset
    population_2 = np.concatenate(([0]*start, [1], [2], [1], [0]*end))
    return population_2 / np.sum(population_2)

population_1 = np.concatenate(([0]*27, [1], [2], [3], [4], [3], [2], [1], [0]*27))
population_1 = population_1 / np.sum(population_1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(range(len(population_1)), population_1, c='k', lw=3, zorder=20, label='Population 1')
for offset in np.arange(-20, 20, 1):
    ax.plot(range(len(population_2(offset))), population_2(offset), alpha=0.5)
ax.legend(frameon=True)
fetching_plot(fig)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
for offset in np.arange(-20, 20, 1):
    this = simulation()
    this.unbound_population = population_1
    this.bound_population = population_2(offset)
    this.simulate(pka_md_data = False)
    ax.scatter(offset, np.mean(this.flux_u + this.flux_b))

ax.set_xlabel('Offset between spikes (bins)')
ax.set_ylabel('Intrasurface flux (cycles/second)')
fetching_plot(fig)
```
