```python
adk_apo_psi   = sorted(glob.glob(base_directory + apo + psi))
surfaces = np.empty((len(adk_apo_psi), 60))
for index, file in enumerate(adk_apo_psi):
    surface = np.genfromtxt(file, delimiter = ' ', skip_header=1, usecols=1)
    surfaces[index] = surface

fig = plt.figure(figsize=(6 * 1.2, 6))
gs = GridSpec(1, 1, wspace=0.2, hspace=0.5)
ax1 = plt.subplot(gs[0, 0])
for surface in surfaces:
    ax1.plot(surface, lw=1)
ax1.set_xticks([0, 15, 30, 45, 60])
ax1.set_xticklabels(
            [r'$-\pi$', r'$-\frac{1}{2}\pi{}$', r'$0$', r'$\frac{1}{2}\pi$', r'$\pi$'])
ax1.set_title('All psi angles')
paper_plot(fig)
```
