```python
data = pd.DataFrame.merge(ATP_binders, Uniprot_PDBs, on='Uniprot Acc')
```

```python
def report_matches(i):
    atp_strings = data[data['Uniprot Acc'] == data['Uniprot Acc'][i]]['PDB ID'].as_matrix()
    search_strings = data[data['Uniprot Acc'] == data['Uniprot Acc'][i]]['PDB'].as_matrix()
    for string in search_strings:
        if string in atp_strings:
            pass
        else:
            print('PDB {} identified by Uniprot ID {} search is not list of ATP-bound PDBs'.format(string, data['Uniprot Acc'][i]))
```

```python
for i in range(len(data)):
    report_matches(i)
```

```python
uniprots_with_structures_in_both = []
for sequence in np.unique(data['Uniprot Acc'].as_matrix()):
    try:
        atp_strings = data[data['Uniprot Acc'] == sequence]['PDB ID'].as_matrix()
        search_strings = data[data['Uniprot Acc'] == sequence]['PDB'].as_matrix()
    except:
        pass
    if len(atp_strings) > 1 and len(search_strings) > 1:
        uniprots_with_structures_in_both.append(sequence)
```
