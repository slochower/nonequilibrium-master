# Can I make a sawtooth wave with a transition period?

```python
def sawtooth_with_transition(down_m, x_down, up_m, x_up, down_offset=0, up_offset=0):
    """
    Return a sawtooth wave as a function of x, with a specified period, and
    transition length as a fraction of the period.
    """
    down_line = down_m * x_down + down_offset
    up_line   = up_m   * x_up   + up_offset
    return np.concatenate((down_line, up_line))
    
```
